```python
import tensorflow as tf
import tensorflow_probability as tfp
from tensorflow.keras.layers import Input, Dense, Concatenate
from tensorflow.keras.models import Model
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/content/drive/MyDrive/eurusd_hour.csv')
change = data['ACh']
change = change.values
print(change)
```

```python
def sliding_window_standardize(arr, window_size=3):

    # Calculate the number of elements to slide the window
    stride = 1

    # Calculate the number of windows and initialize the standardized array
    std_dev = np.std(arr)
    num_windows = arr.shape[0] - window_size+1
    standardized_arr = np.zeros((num_windows, window_size), dtype=float)
    target_arr = np.zeros((num_windows), dtype=float)
    # Loop through the input array with the sliding window
    for i in range(num_windows-1):

        window_slice = arr[i:i+window_size]  # Extract the window slice

        target_arr[i] = arr[i+window_size]/std_dev

        scaler = np.max(np.abs(window_slice))

        standardized_slice = window_slice / scaler

        standardized_arr[i] = standardized_slice

    return standardized_arr, target_arr
```

```python
shifted = np.roll(change, 1)
print(shifted)
```

```python
window_size = 30
window,target = sliding_window_standardize(change,window_size)
np.set_printoptions(suppress=True)
print(window)
print(target)
```

```python
tfd = tfp.distributions
drop_rate = .5
# Define the input layer
inputs = tf.keras.Input(shape=(30,))

# Neural network layers
x = tf.keras.layers.Dense(50, activation='gelu')(inputs)
x = tf.keras.layers.Dropout(drop_rate)(x)
x = tf.keras.layers.Dense(50, activation='gelu')(x)
x = tf.keras.layers.Dropout(drop_rate)(x)
x = tf.keras.layers.Dense(20, activation='gelu')(x)
x = tf.keras.layers.Dropout(drop_rate)(x)
# Store this layer for later use
mean = tf.keras.layers.Dense(1, activation='linear')(x)
vari = tf.keras.layers.Dense(1, activation='relu')(x)

combined = tf.keras.layers.concatenate([mean,vari])
# Create the model
output = tfp.layers.DistributionLambda(lambda t:
                                       tfd.Normal(loc=t[...,0],
                                                  scale=tf.math.softplus(t[...,1])))(combined)
model = tf.keras.Model(inputs=inputs, outputs=output)

def negative_log_likelihood(y_true, y_pred):
  return -tf.reduce_mean(y_pred.log_prob(y_true))

# Compile the model with the NLL loss function
model.compile(optimizer='adam', loss=negative_log_likelihood)


model.summary()
```

```python
# Train the model
history = model.fit(window,target,64,100,validation_split=0.2)
```

```python
# Create a prediction model without the output layer
prediction_model = tf.keras.Model(inputs=inputs, outputs=combined)
pred = prediction_model.predict(window)
```
